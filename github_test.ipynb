```python
import gensim.downloader as api
from gensim.models import KeyedVectors
import numpy as np
import os
import spacy
from sklearn.metrics.pairwise import cosine_similarity 

glove_file_path = r'path/to/the/file'

# Load spacy Model
try:
    nlp = spacy.load("en_core_web_sm")
    print("spacy model loaded successfully.")
except OSError:
    print("spaCy mode not found. Downloading...")

# Function to tokenize text 
def tokenize_text(text):
    """
    Takes a text string, processes it with spaCy, and returns a list of tokens.
    """
    if not isinstance(text, str):
        print(f"Warning: Expected string but got {type(text)}.")
        return []
    doc = nlp(text)
    tokens = [token.text for token in doc]
    return tokens

# Function to load GloVe 
def load_glove_text(file_path):
    print(f"loading GloVe from: {file_path}")
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found at {file_path}")

    vectors = {}
    dimension = -1
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            try:
                parts = line.strip().split()
                word = parts[0]
                vec = np.array([float(x) for x in parts[1:]], dtype=np.float32)

                if dimension == -1:
                    dimension = len(vec)
                elif len(vec) != dimension:
                    print(f"Warning: Inconsistent dimension at line {i+1}. Expected {dimension}, got {len(vec)}. Skipping.")
                    continue

                vectors[word] = vec
            except ValueError:
                print(f"Skipping line {i+1} due to parsing error (likely not a valid vector line): {line.strip()[:50]}...")
            except Exception as e:
                print(f"An error occurred processing line {i+1}: {e}")

    if not vectors:
        raise ValueError("No vectors were loaded. Check file format and content.")

    print(f"Successfully loaded {len(vectors)} vectors with dimension {dimension}.")

    kv = KeyedVectors(vector_size=dimension)
    kv.add_vectors(list(vectors.keys()), list(vectors.values()))
    return kv

# Load the GloVe model
print("Loading GloVe model...")
model = load_glove_text(glove_file_path)
print("GloVe model loaded successfully.")

# Process Movie Titles 
print("Tokenizing movie titles...")
tokenized_titles = data['title'].apply(tokenize_text)
print("Tokenization complete.")

# Calculate the vector for each title
title_vectors = []
processed_sentences = [] 
print("Calculating vectors for each title...")
for i, tokens_list in enumerate(tokenized_titles):
    current_title_vectors = []
    current_sentence_tokens = [] 
    for token in tokens_list:
        if token in model: # Check if token is in GloVe vocabulary
            current_title_vectors.append(model[token])
            current_sentence_tokens.append(token) 
  
    sentence_str = " ".join(current_sentence_tokens)
    
    if current_title_vectors:
        # Calculate the mean vector for the title
        title_vector = np.mean(current_title_vectors, axis=0)
    else:
        title_vector = np.zeros(model.vector_size)

    title_vectors.append(title_vector)
    processed_sentences.append(sentence_str) 

    if i < 5: # print first 5 titles and their corresponding vector
        print(f"\n--- Example {i+1} ---")
        print(f"Original Title (Cleaned): '{sentence_str}'")
        print(f"Vector (first 10 elements): {title_vector[:10]}...") 
        print(f"Vector Shape: {title_vector.shape}")

# Convert the list of title vectors into a NumPy array for easier manipulation
title_vectors_np = np.array(title_vectors)
print(f"\nCalculated {len(title_vectors_np)} title vectors, each with dimension {model.vector_size}.")

if len(processed_sentences) == len(title_vectors_np):
    print(f"Successfully processed and generated vectors for {len(processed_sentences)} titles.")
else:
    print(f"Mismatch: Processed {len(processed_sentences)} sentences but generated {len(title_vectors_np)} vectors.")
```

```python
import numpy as np

def recommend_movies(movie_title: str, movie_titles: list[str],
                     similarity_matrix: np.ndarray,
                     title_vectors_np: np.ndarray, 
                     top_n: int = 10):
    """
    Given a movie title, return its embedding vector, and the top-N similar movies
    (names + their similarity scores).

    Args:
        movie_title (str): The title of the movie to find recommendations for.
        movie_titles (list[str]): A list of all movie titles, in the same order as the similarity matrix.
        similarity_matrix (np.ndarray): The pre-computed all-pairs cosine similarity matrix.
        title_vectors_np (np.ndarray): The array of all movie title vectors.
        top_n (int): The number of recommendations to return.

    Returns:
        dict: A dictionary containing the input movie's info and a list of recommended movies.
              Returns None if the movie_title is not found.
    """
    try:
        # Find the index of the input movie
        idx = movie_titles.index(movie_title)
    except ValueError:
        print(f"Error: Movie '{movie_title}' not found in the dataset.")
        return None 

    sim_scores = similarity_matrix[idx]
    input_movie_vector = title_vectors_np[idx]

    sorted_indices = np.argsort(sim_scores)[::-1]

    # Filter out the input movie itself
    top_indices = [i for i in sorted_indices if i != idx][:top_n]

    recommendations = []
    for i in top_indices:
        recommendations.append({
            "title": movie_titles[i],
            "similarity": float(sim_scores[i]) 
        })

    result = {
        "input_movie": {
            "title": movie_title,
            "vector": input_movie_vector, 
            "index": idx 
        },
        "recommendations": recommendations
    }
    return result
```

```python
input_title = "Inception"# Replace with the movie title you want recommendations for
input_movie = preprocess(input_title)
recommendation_result = recommend_movies(
        movie_title=input_movie,
        movie_titles=movie_titles,
        similarity_matrix=similarity_matrix, 
        title_vectors_np=title_vectors_np,    
        top_n=10
    )

if recommendation_result:
    print(f"--- Input Movie ---")
    print(f"Title: {recommendation_result['input_movie']['title']}")
    print("-" * 20)

    print(f"\n--- Top {len(recommendation_result['recommendations'])} Recommendations ---")
    for rec in recommendation_result['recommendations']:
        print(f"  - Title: {rec['title']}, Cosine Similarity: {rec['similarity']:.4f}")
else:
    print(f"Could not generate recommendations for '{input_movie}'. Please check the title.")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# For a movie that is not already in the dataset
def get_movie_embedding_glove(movie_title, glove_vectors, nlp):
 
    doc = nlp(movie_title.lower())
    vectors = []
    
    for token in doc:
        if token.text in glove_vectors:
            vectors.append(glove_vectors[token.text])
    
    if not vectors:
        return np.zeros(100)  
    
    return np.mean(vectors, axis=0)

def recommend_movies_new_input(
    input_movie_title,     
    movie_titles,           
    title_vectors_np,     
    glove_vectors,          
    nlp,                    
    top_n=10
):

    
    input_vector = get_movie_embedding_glove(input_movie_title, glove_vectors, nlp)
    
    similarities = cosine_similarity([input_vector], title_vectors_np)[0]
    
    sorted_indices = np.argsort(similarities)[::-1]
    
    recommendations = []
    for idx in sorted_indices:
        if movie_titles[idx].lower() == input_movie_title.lower():
            continue
        recommendations.append({
            'title': movie_titles[idx],
            'similarity': float(similarities[idx])
        })
        if len(recommendations) >= top_n:
            break
    
    return {
        'input_movie': {
            'title': input_movie_title,
            'vector': input_vector
        },
        'recommendations': recommendations
    }
```

# Deployment with OOP

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

class MovieRecommender:
    """
    Movie recommender system
    """
    
    def __init__(self, df, description_embeddings, tfidf_matrix, type_encoded):
       
        self.df = df
        self.description_embeddings = description_embeddings
        self.tfidf_matrix = tfidf_matrix
        self.type_encoded = type_encoded
        
        self.weights = {
            'desc': 1.5,
            'tfidf': 0., 
            'type': 0.1
          #  'pairwise': 0.2
        }
    """  
    def set_weights(self, **kwargs):
        
       # Change wights
      # recommender.set_weights(desc=0.5, pairwise=0.3)
        
        for key, value in kwargs.items():
            if key in self.weights:
                self.weights[key] = value
        
        total = sum(self.weights.values())
        for key in self.weights:
            self.weights[key] /= total
    """
    
    def _calculate_pairwise_scores(self, target_features):
      
        pairwise_scores = []
        
        for idx, row in self.df.iterrows():
            score_vector = []
            
            director_sim = 1 if row['director'] == target_features['director'] else 0
            score_vector.append(director_sim)
            
            country_target = set(target_features['country_list'])
            country_row = set(row['country_list'])
            country_sim = 0.5 if len(country_target & country_row) > 0 else 0
            score_vector.append(country_sim)
            
            cast_target = set(target_features['cast_list'])
            cast_row = set(row['cast_list'])
            cast_sim = 0.7 if len(cast_target & cast_row) > 0 else 0
            score_vector.append(cast_sim)
            
            genre_target = set(target_features['genres'])
            genre_row = set(row['genres'])
            genre_sim = 1 if len(genre_target & genre_row) > 0 else 0
            score_vector.append(genre_sim)
            
            pairwise_scores.append(score_vector)
        
        return np.array(pairwise_scores)
    
    def _build_feature_vectors(self, pairwise_scores):
   
        feature_vectors = []
        
        for idx in range(len(self.df)):
            tfidf_row = self.tfidf_matrix[idx].toarray().flatten()
            
            vec = np.concatenate([
                self.description_embeddings[idx],
                tfidf_row,
                self.type_encoded[idx],
                pairwise_scores[idx]
            ])
            feature_vectors.append(vec)
        
        return np.array(feature_vectors)
    
    def find_similar_movies(self, movie_name, top_n=5):
        """        
        Parameters:
        -----------
        movie_name : str
        top_n : int
        
        Returns:
        --------
        list of dict
        """
        target_idx = self.df[self.df['title'] == movie_name].index[0]
        
        target_features = {
            'director': self.df.loc[target_idx, 'director'],
            'country_list': self.df.loc[target_idx, 'country_list'],
            'cast_list': self.df.loc[target_idx, 'cast_list'],
            'genres': self.df.loc[target_idx, 'genres']
        }
        
        pairwise_scores = self._calculate_pairwise_scores(target_features)
        
        feature_vectors = self._build_feature_vectors(pairwise_scores)
        similarity_matrix = cosine_similarity(feature_vectors)
        similar_scores = similarity_matrix[target_idx]
        
        top_indices = np.argsort(similar_scores)[::-1][1:top_n+1]
        
        results = []
        for idx in top_indices:
            results.append({
                'title': self.df.loc[idx, 'title'],
                'similarity': float(similar_scores[idx]),
                'director': self.df.loc[idx, 'director']
            })
        
        return results
    
    def recommend(self, movie_name, top_n=5):
      
        print(f"Movies similar to '{movie_name}':")
        print("="*50)
        
        results = self.find_similar_movies(movie_name, top_n)
        
        for i, movie in enumerate(results, 1):
            print(f"{i}. {movie['title']}")
            print(f"Similarity: {movie['similarity']:.3f}")
          #  print(f"Director: {movie['director']}")
            print()
        
        return results
```

```python
recommender = MovieRecommender(
    df=df,
    description_embeddings=description_embeddings,
    tfidf_matrix=tfidf_matrix,
    type_encoded=type_encoded
)

recommender.recommend("Friends", top_n=10)
```
